```python
import scipy.io
import matplotlib.pyplot as plt

# Load the .mat file
mat_file_path = r"E:\CP Dataset\[20200428] PUMP DATA\VIbration_ALL\impeller_crack_vibration_ALL.mat"
mat_data = scipy.io.loadmat(mat_file_path)

# Access the actual signal data
signal_data = mat_data['concatenated_rows']

# Print shape and type to confirm structure
print("Shape of 'concatenated_rows':", signal_data.shape)
print("Type of elements:", type(signal_data))

# Plot the first signal (adjust if needed after seeing shape)
plt.figure(figsize=(9, 5))
plt.plot(signal_data[0])  # You can try signal_data[0, :] or signal_data[:, 0] depending on shape
#plt.title("First Signal from Impeller Crack Vibration Data")
plt.xlabel("Time")
plt.ylabel("Amplitude")
#plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the first signal with improved style
plt.figure(figsize=(12, 6))

plt.plot(signal_data[0], color='m', linewidth=1.6, alpha=1)
#plt.title("Impeller Crack Vibration Signal", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("Time (s)", fontsize=16, fontweight = "bold")
plt.ylabel("Amplitude (g)", fontsize=16, fontweight = "bold")

# Grid, ticks, and background
#plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Optional: add a border and background
ax = plt.gca()
ax.set_facecolor('#f9f9f9')
for spine in ax.spines.values():
    spine.set_edgecolor('#666666')
    spine.set_linewidth(1.2)

plt.show()
```
